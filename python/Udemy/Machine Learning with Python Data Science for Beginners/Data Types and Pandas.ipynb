### Working with CSV s with Pandas

```python
import pandas as pd
```

```python
# we ar calling the (.read_csv("")) function from the pandas library to READ a csv file from our machine
# within the function you will need to include the PATH of said .csv
# and storing this call within 'df'
df = pd.read_csv("P9-ConsoleDates.csv")

# Calling df will output the .csv within 
df
```

```python
type(df)
```

```python
# .shape will will output the shape of the the .csv, and as you can see it is 30 rows and 5 columns long
df.shape
```

```python
# .columns will print out the names of the .csv columns
df.columns
```

```python
# .head function will output a specific number of rows
# NOTE: The will start at the TOP of the csv
df.head(10)
```

```python
# .tail function will output a specific number of rows
# NOTE: The will start at the BOTTOM of the csv
df.tail(5)
```

```python
# Outputs info of our imported csv
df.info()
```

```python
df.describe()
```
